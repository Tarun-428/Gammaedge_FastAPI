```python
import pandas as pd
```

```python
df = pd.read_json("log_data.json")
df
```

```python
#Error Logs
error_logs = df[df["log_level"] == "ERROR"]
print(error_logs)
```

```python
#Occurrence of each logs
count = df['log_level'].value_counts()
print(count)
```

```python
#Most active user by user_id
most_active_user = df['user_id'].value_counts().idxmax()
print(most_active_user)
```

```python
# Group errors by hour of the day
error_df = df[df["log_level"] == "ERROR"]
hourly_errors = error_df.groupby(error_df['timestamp'].dt.hour)
for i in hourly_errors:
    print(f"Hour :{i}")
```

```python
""""Generate a summary report showing:
Total logs processed
Error rate percentage
Top 5 most common error messages
Time period with most errors"""
print("Total Logs: ",len(df),"Shape of dataframe ",df.shape)

error_per = len(df[df['log_level']=='ERROR'])/len(df['log_level'])*100
print("Error rate percentage : ",error_per)

top_error = df["message"].value_counts().head(5)
print(f"Top 5 error : {top_error}")

peak_time = hourly_errors.size().idxmax()
print(f"Time Period with most errors : {peak_time}")
```
